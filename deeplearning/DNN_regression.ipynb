```python
import pandas as pd
import numpy as np
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
boston_df = pd.read_csv('./datasets/boston_housing.csv' , header=None , delimiter=r"\s+" , names=column_names)
boston_df.head(3)
```

```python
X = boston_df.drop('MEDV' , axis=1 )
y = boston_df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
    
seed= 0 
np.random.seed(seed)

x_tr, x_test, y_tr, y_test = train_test_split(X,y, test_size=0.3, random_state=seed)

model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

model.compile(loss = 'mean_squared_error',
              optimizer='adam')

model.fit(x_tr,y_tr,epochs=1000, batch_size=10)

y_pred = model.predict(x_test).flatten()

for i in range(10):
    label = y_test[i]
    prediction = y_pred[i]
    print(f'{label:,3f}, :{prediction:,3f}')
```
