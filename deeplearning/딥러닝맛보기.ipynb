```python
# 딥러닝 구동하는데 필요한 케라스 함수를 불러옴
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

#실행할때마다 같은 결과를 출력하기 위해 설정하는 부분
np.random.seed(3)
tf.random.set_seed(3)

#준비된 수술 환자 데이터를 불러옴
data_set = np.loadtxt('./datasets/ThoraricSurgery.csv', delimiter=",")

#환자의 기록과 수술 결과를 x,y 구분하여 저장
x = data_set[:,0:17]
y = data_set[:,17]

#딥러닝 구조를 결정(모델 설정, 실행)
model = Sequential()
model.add(Dense(30, input_dim=17, activation = 'relu'))
model.add(Dense(1, activation='sigmoid'))

#딥러닝 실행
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x,y, epochs=100, batch_size=100)
```

```python
# 딥러닝 구동하는데 필요한 케라스 함수를 불러옴
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
X= data = np.linspace(1,2,200)
y= X*4 + np.random.randn(200)*0.3

model= Sequential()
model.add(Dense(1, input_dim=1, activation= 'linear'))
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.fit(X,y,batch_size=1, epochs=30)

predict = model.predict(data)
plt.plot(data, predict, 'b',data, y, 'k.' )
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X= data = X.reshape(-1,1)
lr = LinearRegression()
lr.fit(X,y)
pred = lr.predict(data)
plt.plot(data, pred, 'b',data, y, 'k.' )
plt.show()
```
