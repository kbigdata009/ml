```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
import tensorflow as tf 
from keras.layers import Dense
#from keras.layers.core import Dense 

df=pd.read_csv('./datasets/sonar.csv',header=None)

import numpy as np 
np.random.seed(3)
tf.random.set_seed(3)

data = df.values 
x=data[:,:60]
x=np.asarray(x).astype(np.float32)

y=data[:,60]

e= LabelEncoder()
e.fit(y)
y= e.transform(y)

model= Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(x,y,epochs=200, batch_size=5)
#배치 수 작아지면 과적합. 

print(f'정확도:{model.evaluate(x,y)[1]}')
```

```python
from sklearn.model_selection import StratifiedKFold 

n_fold= 10
seed=3
kfold = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)

accuracy=[]

for train, test in kfold.split(x,y): #kfold를 10번 접었으므로 10번 반복중. 
    model= Sequential()
    model.add(Dense(24,input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])
    
    model.fit(x[train], y[train], epochs=100, batch_size=5)
    k_accuracy=round(model.evaluate(x[test],y[test])[1],3)
    accuracy.append(k_accuracy)
```
